```python
import pandas as pd

df = pd.read_csv('~/Downloads/day3.txt', header=None, names=['items'])
df
```

```python
df['items_in_both'] = df['items'].apply(lambda x: set(x[:int(len(x)/2)]).intersection(set(x[int(len(x)/2):])))
df
```

```python
df['priorities'] = df['items_in_both'].apply(lambda x: sum([ord(c) - (38 if c.isupper() else 96) for c in x]))
df
```

```python
df.priorities.sum()
```

```python
df2 = df.groupby(df.index // 3).items.apply(list).reset_index()
df2
```

```python
df2['common_item_in_group'] = df2['items'].apply(lambda x: set.intersection(*[set(l) for l in x]))
df2
```

```python
df2['priorities2'] = df2['common_item_in_group'].apply(lambda x: sum([ord(c) - (38 if c.isupper() else 96) for c in x]))
df2.priorities2.sum()
```
