```python
viewing_distance = np.zeros((len(df), len(df)))
```

```python
df.loc[:,0] = 9
df.loc[:,len(df)-1] = 9
df.loc[0] = 9
df.loc[len(df)-1] = 9
df
```

```python
def while_true(l):
    i = 0
    while i < len(l) and l[i]:
        yield l[i]
        i += 1

for r in range(1, len(df) - 1):
    for c in range(1, len(df) - 1): 
        row = list(df.loc[r].apply(lambda x: x < df.loc[r ,c]))
        col = list(df.loc[:,c].apply(lambda x: x < df.loc[r ,c]))
        d_left = sum([x for x in while_true(row[:c][::-1])]) + 1
        d_right = sum([x for x in while_true(row[c+1:])]) + 1
        d_up = sum([x for x in while_true(col[:r][::-1])]) + 1
        d_down = sum([x for x in while_true(col[r+1:])]) + 1
        viewing_distance[r, c] = d_left * d_right * d_up * d_down
np.max(viewing_distance)
```
