```python
import pandas as pd
import numpy as np

df = pd.read_csv('~/Downloads/day2.txt', header=None, delimiter=' ', names=['You', 'Me'])
df
```

```python
outcomes = {
    ('A','X'): 'draw',
    ('A','Y'): 'won',
    ('A','Z'): 'lost',
    ('B','X'): 'lost',
    ('B','Y'): 'draw',
    ('B','Z'): 'won',
    ('C','X'): 'won',
    ('C','Y'): 'lost',
    ('C','Z'): 'draw'
}
shape_score = {
    'X': 1,
    'Y': 2,
    'Z': 3
}
outcome_score = {
    'lost': 0,
    'draw': 3,
    'won': 6
}
```

```python
df['outcome'] = df.apply(lambda x: outcomes[(x.You, x.Me)], axis=1)
df
```

```python
df['score'] = df.apply(lambda x: outcome_score[x.outcome] + shape_score[x.Me], axis=1)
df
```
