```python
import pandas as pd
import numpy as np

df = pd.read_csv('~/Downloads/day2.txt', header=None, delimiter=' ', names=['You', 'Me'])
df
```

```python
outcomes = {
    ('A','X'): 'draw',
    ('A','Y'): 'won',
    ('A','Z'): 'lost',
    ('B','X'): 'lost',
    ('B','Y'): 'draw',
    ('B','Z'): 'won',
    ('C','X'): 'won',
    ('C','Y'): 'lost',
    ('C','Z'): 'draw'
}
shape_score = {
    'X': 1,
    'Y': 2,
    'Z': 3
}
outcome_score = {
    'lost': 0,
    'draw': 3,
    'won': 6
}
```

```python
df['outcome'] = df.apply(lambda x: outcomes[(x.You, x.Me)], axis=1)
df
```

```python
df['score'] = df.apply(lambda x: outcome_score[x.outcome] + shape_score[x.Me], axis=1)
df
```

```python
df.score.sum()
```

```python
desired_outcome = {
    'lost': 'X',
    'draw': 'Y',
    'won': 'Z'
}
desired_outcome_reverse = {v:k for k,v in desired_outcome.items()}
reverse_lookup = {(k[0], desired_outcome[v]): k[1] for k,v in outcomes.items()}
reverse_lookup
```

```python
df['Me_2'] = df.apply(lambda x: reverse_lookup[(x.You, x.Me)], axis=1)
df
```

```python
df['score_2'] = df.apply(lambda x: outcome_score[desired_outcome_reverse[x.Me]] + shape_score[x.Me_2], axis=1)
df.score_2.sum()
```
